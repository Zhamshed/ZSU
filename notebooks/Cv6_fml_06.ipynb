```python
X = df_scaled.values

ks = range(2, 9)
inertias = []
sil_scores = []

for k in ks:
    km = sklearn.cluster.KMeans(n_clusters=k, random_state=42, n_init="auto")
    labels = km.fit_predict(X)
    inertias.append(km.inertia_)
    sil_scores.append(sklearn.metrics.silhouette_score(X, labels))

plt.figure(figsize=(6,4))
plt.plot(list(ks), inertias, marker='o'); plt.title("Elbow (Inertia)")
plt.xlabel("k"); plt.ylabel("Inertia"); plt.tight_layout(); plt.show()

plt.figure(figsize=(6,4))
plt.plot(list(ks), sil_scores, marker='o'); plt.title("Silhouette vs k")
plt.xlabel("k"); plt.ylabel("Silhouette"); plt.tight_layout(); plt.show()

best_k = list(ks)[int(np.argmax(sil_scores))]  #k with max
best_k
```

```python
kmeans = sklearn.cluster.KMeans(n_clusters=best_k, random_state=42, n_init="auto")
labels = kmeans.fit_predict(X)

df_with_clusters = df.copy()
df_with_clusters["cluster"] = labels

pd.Series(labels).value_counts().sort_index()
```

```python
#2 Perform an exploration analysis of the clustered data
X_scaled_with_c = df_scaled.copy()
X_scaled_with_c["cluster"] = labels
cluster_profile = X_scaled_with_c.groupby("cluster").mean().round(3)
cluster_profile
```

```python
pca = sklearn.decomposition.PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(6,5))
for c in range(best_k):
    m = (labels == c)
    plt.scatter(X_pca[m, 0], X_pca[m, 1], s=30, alpha=0.85, label=f"C{c}")
plt.title("Clusters in PCA space (2D)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.legend(); plt.tight_layout(); plt.show()
```

```python
#3
for c in range(best_k):
    means = cluster_profile.loc[c].sort_values(ascending=False)
    top3 = ", ".join([f"{f}~{v:.2f}" for f, v in means.head(3).items()])
    low3 = ", ".join([f"{f}~{v:.2f}" for f, v in means.tail(3).items()])
    print(f"Cluster {c} | size={sum(labels==c)} | high: {top3} | low: {low3}")
```
