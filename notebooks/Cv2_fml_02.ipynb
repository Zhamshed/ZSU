## âœ… Task 1 - Outlier detection (1p)
- We need to somehow mark the outliers in the data according to the *SalePrice* and *GrLivArea*
    - One possibility is to compute [IQR](https://www.statisticshowto.com/probability-and-statistics/interquartile-range/) for both columns and mark outliers using lower and upper bounds
    - Lower bound: Q1 - 1.5*IQR
    - Upper bound: Q3 + 1.5*IQR
- ðŸ’¡ If the house has *SalePrice* **OR** *GrLivArea* value outside of the bounds - it is an outlier
- Vizualize the data using scatter plot and distinguish the outlier and non-outlier data using different colors (*hue*) ðŸ’¡

## âœ… Task 2 - Describe what you see in the data (1p)
- Try to vizualize the relationship between *SalePrice* and *OverallQual*
    - ðŸ’¡You can use BoxPlots, Scatter plots, etc., the choice of a right plot type is up to you ðŸ™‚
- Do the same for *SalePrice* and *OverallCond*; i.e. vizualize and describe insight-
- **Describe the insight you got from the plots with a few sentences in a Markdown cell below the plot**
    - âŒ Plot interpretation figured in real-time during task check is not allowed! âŒ

```python
#task1
def iqr_bounds(s: pd.Series):
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return lower, upper

sp_low, sp_up = iqr_bounds(df['SalePrice'])
gl_low, gl_up = iqr_bounds(df['GrLivArea'])

out_sp = (df['SalePrice'] < sp_low) | (df['SalePrice'] > sp_up)
out_gl = (df['GrLivArea'] < gl_low) | (df['GrLivArea'] > gl_up)
df['IsOutlier'] = out_sp | out_gl

print(f"IQR bounds â€” SalePrice: [{sp_low:,.0f}, {sp_up:,.0f}] | GrLivArea: [{gl_low:,.0f}, {gl_up:,.0f}]")
print(f"Outliers flagged: {df['IsOutlier'].sum()} / {len(df)}")

plt.figure(figsize=(9,6))
ax = sns.scatterplot(
    data=df, x='GrLivArea', y='SalePrice',
    hue='IsOutlier', alpha=0.8
)
ax.set_title('SalePrice vs GrLivArea (IQR Outliers Highlighted)', pad=12)
ax.set_xlabel('GrLivArea (Above-Ground Living Area)')
ax.set_ylabel('SalePrice')
plt.tight_layout()
plt.show()
```

```python
#task2
plt.figure(figsize=(9,6))
sns.boxplot(
    data=df,
    x='OverallQual', y='SalePrice',
    order=sorted(df['OverallQual'].dropna().unique())
)
plt.title('SalePrice by OverallQual (Boxplot)')
plt.tight_layout()
plt.show()

med_qual = df.groupby('OverallQual', as_index=False)['SalePrice'].median().sort_values('OverallQual')
plt.figure(figsize=(9,6))
sns.pointplot(data=med_qual, x='OverallQual', y='SalePrice')
plt.title('Median SalePrice by OverallQual')
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,6))
sns.boxplot(
    data=df,
    x='OverallCond', y='SalePrice',
    order=sorted(df['OverallCond'].dropna().unique())
)
plt.title('SalePrice by OverallCond (Boxplot)')
plt.tight_layout()
plt.show()

med_cond = df.groupby('OverallCond', as_index=False)['SalePrice'].median().sort_values('OverallCond')
plt.figure(figsize=(9,6))
sns.pointplot(data=med_cond, x='OverallCond', y='SalePrice')
plt.title('Median SalePrice by OverallCond')
plt.tight_layout()
plt.show()
```
